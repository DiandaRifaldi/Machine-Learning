```python
import pandas as pd
```

```python
data = pd.read_csv("heart_attack_dataset.csv")
data
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.drop(['total Has Diabetes', 'total Smoking Status', 'total Chest Pain Type', 'total Treatment'], 
          axis='columns', inplace=True)
```

```python
data
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt

# Statistik deskriptif untuk variabel numerik
numerical_summary = data[['Age', 'Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']].describe()
print(numerical_summary)

# Histogram untuk distribusi usia, tekanan darah, dan kolesterol
data[['Age', 'Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']].hist(bins=20, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
# Proporsi pasien berdasarkan Gender
gender_proportion = data['Gender'].value_counts(normalize=True)
print("Proporsi Gender:")
print(gender_proportion)

# Proporsi pasien berdasarkan Diabetes
diabetes_proportion = data['Has Diabetes'].value_counts(normalize=True)
print("\nProporsi Diabetes:")
print(diabetes_proportion)

# Proporsi pasien berdasarkan Kebiasaan Merokok
smoking_proportion = data['Smoking Status'].value_counts(normalize=True)
print("\nProporsi Kebiasaan Merokok:")
print(smoking_proportion)

# Visualisasi proporsi dengan bar chart
gender_proportion.plot(kind='bar', title='Proporsi Gender')
plt.show()

diabetes_proportion.plot(kind='bar', title='Proporsi Diabetes')
plt.show()

smoking_proportion.plot(kind='bar', title='Proporsi Kebiasaan Merokok')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram untuk distribusi usia
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribusi Usia Pasien')
plt.xlabel('Usia (tahun)')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Boxplot untuk tekanan darah berdasarkan jenis pengobatan
plt.figure(figsize=(10, 6))
sns.boxplot(x='Treatment', y='Blood Pressure (mmHg)', data=data, palette='Set2')
plt.title('Distribusi Tekanan Darah Berdasarkan Jenis Pengobatan')
plt.xlabel('Jenis Pengobatan')
plt.ylabel('Tekanan Darah (mmHg)')
plt.show()

# Boxplot untuk kolesterol berdasarkan jenis pengobatan
plt.figure(figsize=(10, 6))
sns.boxplot(x='Treatment', y='Cholesterol (mg/dL)', data=data, palette='Set2')
plt.title('Distribusi Kolesterol Berdasarkan Jenis Pengobatan')
plt.xlabel('Jenis Pengobatan')
plt.ylabel('Kolesterol (mg/dL)')
plt.show()
```

```python
# Bar chart untuk jenis nyeri dada
plt.figure(figsize=(10, 6))
data['Chest Pain Type'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribusi Jenis Nyeri Dada')
plt.xlabel('Jenis Nyeri Dada')
plt.ylabel('Frekuensi')
plt.xticks(rotation=0)
plt.show()
```

```python
# Menghitung korelasi antar kolom numerik
correlation_matrix = data[['Age', 'Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']].corr()

# Visualisasi korelasi dengan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Korelasi Antar Usia, Tekanan Darah, dan Kolesterol')
plt.show()
```

```python
# Boxplot untuk kolesterol berdasarkan gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Cholesterol (mg/dL)', data=data, palette='Set1')
plt.title('Distribusi Kolesterol Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Kolesterol (mg/dL)')
plt.show()
```

```python
# Countplot untuk jenis nyeri dada berdasarkan jenis pengobatan
plt.figure(figsize=(10, 6))
sns.countplot(x='Treatment', hue='Chest Pain Type', data=data, palette='Set2')
plt.title('Jenis Nyeri Dada Berdasarkan Pengobatan')
plt.xlabel('Jenis Pengobatan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Proporsi pasien berdasarkan merokok
smoking_proportion = data['Smoking Status'].value_counts(normalize=True)

# Bar chart untuk merokok
plt.figure(figsize=(8, 6))
smoking_proportion.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Proporsi Pasien Berdasarkan Merokok')
plt.xlabel('Kebiasaan Merokok')
plt.ylabel('Proporsi')
plt.xticks(rotation=0)
plt.show()

# Proporsi pasien berdasarkan diabetes
diabetes_proportion = data['Has Diabetes'].value_counts(normalize=True)

# Bar chart untuk diabetes
plt.figure(figsize=(8, 6))
diabetes_proportion.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Proporsi Pasien Berdasarkan Diabetes')
plt.xlabel('Status Diabetes')
plt.ylabel('Proporsi')
plt.xticks(rotation=0)
plt.show()
```

```python
# Countplot untuk jenis pengobatan berdasarkan gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Treatment', hue='Gender', data=data, palette='Set2')
plt.title('Proporsi Jenis Pengobatan Berdasarkan Gender')
plt.xlabel('Jenis Pengobatan')
plt.ylabel('Jumlah Pasien')
plt.legend(title='Gender')
plt.show()
```

```python
# Membuat kolom baru untuk kelompok usia
bins = [30, 40, 60, 80]  # Tentukan batas usia
labels = ['<40', '40-60', '60+']  # Label untuk kelompok usia
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels)

# Countplot untuk jenis pengobatan berdasarkan kelompok usia
plt.figure(figsize=(10, 6))
sns.countplot(x='Treatment', hue='Age Group', data=data, palette='Set2')
plt.title('Proporsi Jenis Pengobatan Berdasarkan Kelompok Usia')
plt.xlabel('Jenis Pengobatan')
plt.ylabel('Jumlah Pasien')
plt.legend(title='Kelompok Usia')
plt.show()
```

```python
# Countplot untuk jenis nyeri dada
plt.figure(figsize=(10, 6))
sns.countplot(x='Chest Pain Type', data=data, palette='Set2')
plt.title('Distribusi Jenis Nyeri Dada')
plt.xlabel('Jenis Nyeri Dada')
plt.ylabel('Jumlah Pasien')
plt.show()
```
